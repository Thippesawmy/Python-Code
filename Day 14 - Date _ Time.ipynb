```python
# Timedelta

from datetime import date as D, datetime as dt
```

```python
D1 = D(2020, 1, 1)
D2 = D(2020, 9, 27)
print('Difference between Date 1 & 2 is:', D2 - D1)
```

```python
D1 = dt(2020, 1, 1, 12, 0, 0)
D2 = dt(2020, 9, 27, 10, 20, 54)
print('Difference between Date 1 & 2 is:', D2 - D1)
```

```python
from datetime import timedelta as TD
```

```python
t1 = TD(seconds = 45)
t2 = TD(seconds = 54)
print('Difference between T1 & T2 in Seconds is', t2-t1)
```

```python
now = dt.now()
print(now)

futuredays1 = now + TD(days = 730) #After 2 Years
futuredays2 = now + TD(days = 9)

print('After 2 years, the Date will be', futuredays1)
print('After 9 Days, the Date will be', futuredays2)
```

```python
t3 = TD(days = 5, hours = 12, seconds = 45, microseconds = 132456)
print(t3.total_seconds())
```

```python
# Formatting - Strftime()

now = dt.now()
print('Current Time is', now)

print('---------------------')

a = now.strftime("%m-%d-%y, %H:%M:%S")
print('Formatted Date is', a)
```

```python
now = dt.now()
print('Current Time is', now)

a = now.strftime("%m/%d/%Y, %H:%M:%S")
print(a)

b = now.strftime("%d %b, %Y")
print(b)

c = now.strftime("%d %B, %Y")
print(c)

d = now.strftime("%I %a")
print(d)

e = now.strftime("%I %A")
print(e)

f = now.strftime("%H:%M %p")
print(f)
```

```python
# Strptime = COnvert a string into a Date

a = '27 September 2020'
print('String Date is', a)

D1 = dt.strptime(a, "%d %B %Y")
print(D1)
```

```python
# Handling Timezones

local = dt.now()
print('Local Time is', local)
```

```python
import pytz

tznew = pytz.timezone("Australia/Sydney")
a = dt.now(tznew)
print('Current time in Sydney is', a.strftime('%d %B %Y, %H:%M %p'))
```

```python
tznew = pytz.timezone("Europe/Paris")
a = dt.now(tznew)
print('Current time in Paris is', a.strftime('%d %B %Y, %H:%M %p'))
```

```python
# Sleep

import time as T

print('This will get Executed First')
T.sleep(5)
print('This is Executed after 5 Seconds')
```

```python
# Watch

while True:
    local = T.localtime()
    a = T.strftime("%H:%M:%S %p", local)
    print(a, end = '')
    print('\r', end = '')
    T.sleep(1)
```
